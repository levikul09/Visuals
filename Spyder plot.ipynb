```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/kaggle/input/top-650-fifa-22-players-simplified/Top_650_FIFA.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
vzdf= df[['short_name','shooting', 'passing', 'dribbling','skill', 'movement', 'power', 'mentality']].iloc[0:3,:]
```

```python
vzdf
```

```python
fig = go.Figure()

for i, row in vzdf.iterrows():
    fig.add_trace(go.Scatterpolar(
      r=vzdf.iloc[i, 1:8].values,
      theta=vzdf.columns.difference(['short_name']),
      fill='toself',
        name=vzdf.iloc[i,0]
    ))


fig.update_layout(
    width=1600,
    height=900,
    font=dict(size=26),
    showlegend=True,
    polar=dict(
        radialaxis=dict(
        visible=True,
        range=[65, 100]
            ))
  )

fig.show()
```
